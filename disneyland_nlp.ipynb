 Importing libraries

```python
import numpy as np
import pandas as pd
import nltk
import string
```

```python
nltk.download('all')
```

```python
dlrev=pd.read_csv('/content/DisneylandReviews.csv', encoding='latin-1')
dlrev.head()
```

```python
dlrev.shape
```

```python
df=dlrev[['Rating','Reviewer_Location','Review_Text','Branch']]
df.head()
```

# Removing Punctuations

```python
def remove_punc(sentence):
    text="".join([i for i in sentence if i not in string.punctuation])
    return text

df['text']=df['Review_Text'].apply(lambda x:remove_punc(x))
df.head()
```

# Lower casing sentences

```python
df['text']=df['text'].apply(lambda x:x.lower())
df.head()
```

# Removing Stopwords

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['text'] = df['text'].apply(lambda x : " ".join(x for x in x.split() if x not in stop))
df.head()
```

# Tokenization

```python
from nltk.tokenize import word_tokenize
def tokenize(sentence):
    words=word_tokenize(sentence)
    return words
df['word']=df['text'].apply(lambda x:tokenize(x))
df.head()
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stemming(list):
    stem_text = [ps.stem(word) for word in list]
    return stem_text

df['stem']=df['word'].apply(lambda x:stemming(x))
df.head()
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer 
wl=WordNetLemmatizer()

def lemmatiz(list):
    lem_text=[wl.lemmatize(word) for word in list]
    return lem_text


df['lem']=df['word'].apply(lambda x:lemmatiz(x))
df.head()
```

# Preparing data

```python
def listToString(s):
    str1 = " "
    return (str1.join(s))
      
df['final_text']=df['lem'].apply(lambda s:listToString(s))
df.head()
```

```python
df1=df.drop(['text','word','stem','Review_Text'], axis=1)
df1.head()
```
